```python
#load datasets
dataset=pd.read_csv('Iris.csv')
Id=dataset['Id']
SepalLengthCm=dataset['SepalLengthCm']
SepalWidthCm=dataset['SepalWidthCm']
PetalLengthCm=dataset['PetalLengthCm']
PetalWidthCm=dataset['PetalWidthCm']
dataset.head()
```

```python
# plotting Histogram
plt.hist(SepalLengthCm,bins=10,alpha=0.9,rwidth=0.8,color='blue',label='SepalLengthCm')
plt.hist(SepalWidthCm,bins=10,alpha=0.9,rwidth=0.8,color='orange',label='SepalWidthCm')
plt.hist(PetalLengthCm,bins=10,alpha=0.9,rwidth=0.8,color='green',label='PetalLengthCm')
plt.hist(PetalWidthCm,bins=10,alpha=0.9,rwidth=0.8,color='yellow',label='PetalWidthCm')
plt.legend()
```

```python
plt.figure(figsize=(5,5))
s=data['SepalLengthCm']
plt.hist(s,bins=25,color='blue',alpha=0.7)
plt.title('Sepal Length in Cm')
plt.xlabel('sepal_Length_Cm')
plt.ylabel('count')
```

```python
plt.figure(figsize=(5,5))
t=data['SepalWidthCm']
plt.hist(t,bins=25,color='orange',alpha=0.4)
plt.title('Sepal Width in Cm')
plt.xlabel('sepal_Width_Cm')
plt.ylabel('count')
```

```python
plt.figure(figsize=(5,5))
t=data['PetalLengthCm']
plt.hist(t,bins=25,color='green',alpha=0.9)
plt.title('Petal Length in Cm')
plt.xlabel('Petal_Length_Cm')
plt.ylabel('count')
```

```python
plt.figure(figsize=(5,5))
t=data['PetalWidthCm']
plt.hist(t,bins= 25,color= 'yellow',alpha= 0.8)
plt.title('Petal Width in Cm')
plt.xlabel('Petal_Width_Cm')
plt.ylabel('count')
```
